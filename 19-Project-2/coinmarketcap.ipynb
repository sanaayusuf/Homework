```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
url="https://coinmarketcap.com/currencies/ripple/historical-data/?start=20130428&end=20190821"
```

```python
#webscraping
tables = pd.read_html(url)
tables
```

```python
#createdataframe
df = tables[0]
df.columns = ["Date","Open","High","Low","Close","Volume","Market Cap"]
df.head()
```

```python
#cleanupdata
df = df.iloc[2:]
df.head()
```

```python
df.set_index('Date', inplace=True)
df.head()
```

```python
df['Low'].min()
```

```python
df['High'].max()
```

```python
engine = create_engine("postgresql+psycopg2://postgres:password@localhost/crypto")
```

```python
df.to_sql("crypto",engine,if_exists='replace',index='date')
```

```python
df2=pd.read_sql("""Select * from crypto
                          """,engine)
```
