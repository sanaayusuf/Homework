```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"





```

```python
# Read the City and Ride Data
city_data_df=pd.read_csv("data/city_data.csv")
ride_data_df=pd.read_csv("data/ride_data.csv")

city_data_df
ride_data_df
```

```python
# Combine the data into a single dataset
# Display the data table for preview

merged_df=city_data_df.merge(ride_data_df,on='city', how='inner')
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
by_city=merged_df.groupby('city')
#average fare per city 
avg_fare_city=by_city.mean()['fare']
#total number of rides per city
total_rides_city=by_city.count()['ride_id']
#total number of drivers per city 
total_drivers_city=by_city.mean()['driver_count']
#city type(urban,suburban,rural)
city_type=city_data_df.set_index('city')['type']

#DF 

city_print=pd.DataFrame({
    'Number of Rides':total_rides_city, 
    'Average Fare':avg_fare_city,
    'Total Number of Drivers per City':total_drivers_city,
    'Type of City': city_type
}) 

city_print

city_print.sort_values('Total Number of Drivers per City', ascending=False)
```

```python
# Obtain the x and y coordinates for each of the three city types
Urban=city_print[city_print['Type of City']=='Urban']
Suburban=city_print[city_print['Type of City']=='Suburban']
Rural=city_print[city_print['Type of City']=='Rural']


#colorscheme

city_color={'Urban':'Coral','Suburban':'SkyBlue', 'Rural':'Gold'}


# Build the scatter plots for each city types
# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure


plt.scatter(Urban['Number of Rides'], Urban['Average Fare'], \
s=Urban['Total Number of Drivers per City']*10, color=city_color['Urban'],edgecolor='black', \
            label='Urban', alpha = .75)
plt.scatter(Suburban['Number of Rides'], Suburban['Average Fare'], \
            s = Suburban['Total Number of Drivers per City']*10, color=city_color['Suburban'], edgecolor='black', \
            label='Suburban', alpha=.75)
plt.scatter(Rural['Number of Rides'], Rural['Average Fare'], \
            s = Rural['Total Number of Drivers per City']*10, color=city_color['Rural'], edgecolor='black', \
            label='Rural', alpha=.75)








plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($) Per City')
lgnd = plt.legend(frameon=True, edgecolor='black', title='City Types')
lgnd.legendHandles[0]._sizes = [60]
lgnd.legendHandles[1]._sizes = [60]
lgnd.legendHandles[2]._sizes = [60]
                      
                      
                      
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_groupped = merged_df.groupby(['type'])
fare_sum = type_groupped['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(fare_sum, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python

# Calculate Type Percents
rides_count = type_groupped['ride_id'].count()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(rides_count, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Type Percents
type_groupped_drivers = merged_df.groupby(['type'])
drivers_sum = type_groupped_drivers['driver_count'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(drivers_sum, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")

# Show Figure
plt.show()
```
