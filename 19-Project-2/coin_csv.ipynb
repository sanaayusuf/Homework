```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "consolidated_coin_data.csv"
coin_data_df= pd.read_csv(csv_file)
coin_data_df.head()
```

```python
coin_data_df.set_index('Date', inplace=True)
coin_data_df.head()
```

```python
coin_data_df['Low'].min()
```

```python
coin_data_df['High'].max()
```

```python
engine = create_engine("postgresql+psycopg2://postgres:password@localhost/crypto")
coin_data_df.to_sql("crypto2",engine,if_exists='replace',index='date')
```
